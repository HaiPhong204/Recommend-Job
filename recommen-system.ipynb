```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer, WordNetLemmatizer
import nltk
from sklearn.metrics import classification_report, accuracy_score
user_data = pd.read_excel('user.xlsx')
job_data = pd.read_excel('job.xlsx')
```

```python
cols = ['Name', 'Title', 'Skill', "Experiences"]
users = user_data[cols]
users.head()
```

```python
cols = ['Job Name', 'Label', 'Description', 'Requirement']		
jobs = job_data[cols]
jobs.head()
```

```python
from pre_process_data import stopWords_Teencode
users["Skill"] = users["Skill"].apply(stopWords_Teencode)
users["Experiences"] = users["Experiences"].apply(stopWords_Teencode)
users.head()
```

```python
jobs["Description"] = jobs["Description"].apply(stopWords_Teencode)
jobs["Requirement"] = jobs["Requirement"].apply(stopWords_Teencode)
jobs.head()
```

```python
users['Title'].value_counts()
```

```python
users["Skill"] = users["Skill"].str.replace('\n', ' ')
users["Experiences"] = users["Experiences"].str.replace('\n', ' ')
jobs["Description"] = jobs["Description"].str.replace('\n', ' ')
jobs["Requirement"] = jobs["Requirement"].str.replace('\n', ' ')
```

```python
import string
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree
#storing the puntuation free text
users["Skill"] = users["Skill"].apply(lambda x:remove_punctuation(x))
users["Experiences"] = users["Experiences"].apply(lambda x:remove_punctuation(x))
jobs["Description"] = jobs["Description"].apply(lambda x:remove_punctuation(x))
jobs["Requirement"] = jobs["Requirement"].apply(lambda x:remove_punctuation(x))
```

```python
users.head()
```

```python
jobs.head()
```

```python
users["inform"] = users["Skill"] + users["Experiences"]
jobs["inform"] = jobs["Description"] + jobs["Requirement"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from sklearn.metrics.pairwise import cosine_similarity
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix_skills = tfidf_vectorizer.fit_transform(users["inform"])
tfidf_matrix_descriptions = tfidf_vectorizer.transform(jobs["inform"])
```

```python
cosine_sim = cosine_similarity(tfidf_matrix_skills, tfidf_matrix_descriptions)
cosine_sim
```

```python
user_index = 0
print(users.iloc[user_index])
cosine_sim_user = cosine_sim[user_index]

# Xếp hạng (ranking) các công việc dựa trên độ tương đồng cosine similarity
job_ranking = np.argsort(cosine_sim_user)[::-1]  # Sắp xếp giảm dần
recommended_jobs = jobs.iloc[job_ranking]

# In ra danh sách các công việc được đề xuất
print(recommended_jobs.head(10))
```
